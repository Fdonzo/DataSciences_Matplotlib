## Matplotlib work example

```python
# ste_1: import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#step_2: prepare data

x =[1,2,4,7,8]
y=[3,8,9,13,19]

# step_3: set up  plot
fig, ax =plt.subplots(figsize=(10,10))

# step_4: plot data
ax.plot(x,y)

# step_4: customise plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# step_5: save the plot
fig.savefig("./sample_image.png")
```

```python
#Two options to create subplots
# subplots option 1: 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# plotting different plots
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# plotting different plots
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

## Which one to use: pyplot method or OO method
- when you are using something quick, use pyplot method
- when you are using something advanced, use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
over_50.head()
```

```python
len(over_50)
```

```python
# pyplot method
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
# OO method mixed with pyplot
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target",ax=ax);
ax.set_xlim([45,100])
```

```python
ax.set_xlim([45,100])
```

```python
# OO method from scratch

fix, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# customize the plot
ax.set(title="Heart Disease and Cholesteral Levels", xlabel="Age",ylabel="Cholesteral")

# add legend
ax.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax.axhline(over_50["chol"].mean(),ls="--");
```

```python
over_50.head()
```

```python
# OO method

# define plot variables
fig,(ax0, ax1) = plt.subplots(nrows=2, ncols=1,figsize=(10,10))

#add to data plot
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
scatter1 = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])

# customise plot
ax0.set(title="Heart Disease and Cholesteral Levels", xlabel="Age",ylabel="Cholesteral");
ax1.set(title="Heart Disease and Cholesteral Levels", xlabel="Age",ylabel="Max Heart Rate");

# add legend
ax0.legend(*scatter.legend_elements(), title="Target")
ax1.legend(*scatter.legend_elements(), title="Target")

# add mean line
ax0.axhline(over_50["chol"].mean(),ls="-.")
ax1.axhline(over_50["thalach"].mean(),ls="--");
```

```python

# OO method 
# To share X axes

# define plot variables
fig,(ax0, ax1) = plt.subplots(nrows=2, ncols=1,figsize=(10,10), sharex=True)

#add to data plot
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
scatter1 = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])

# customise plot
ax0.set(title="Heart Disease and Cholesteral Levels",ylabel="Cholesteral");
ax1.set(title="Heart Disease and Cholesteral Levels",xlabel="Age",ylabel="Max Heart Rate");

# add legend
ax0.legend(*scatter.legend_elements(), title="Target")
ax1.legend(*scatter.legend_elements(), title="Target")

# add mean line
ax0.axhline(over_50["chol"].mean(),ls="-.")
ax1.axhline(over_50["thalach"].mean(),ls="--")

# add figure title
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
